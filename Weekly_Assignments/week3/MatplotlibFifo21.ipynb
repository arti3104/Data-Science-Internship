```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load dataset
```

```python
df = pd.read_csv('players_21.csv')
df.head()
```

```python
#Initial Overview
```

```python
print("Rows,Cols:", df.shape)
df.info()
df.describe()
```

```python
#Missing & Duplicates
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
print("Duplicates:", df.duplicated().sum())
```

```python
# Cleaning — drop unused cols, fill NAs
```

```python
df = df.drop(['sofifa_id','player_url'], axis=1)  # adjust as needed
df['pace'] = df['pace'].fillna(df['pace'].median())
```

```python
#Univariate Analysis
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['overall'], bins=30, kde=True)
plt.title('Distribution of Player Overall Ratings')
plt.show()
plt.figure(figsize=(10,5))
sns.countplot(y='nationality', data=df, order=df['nationality'].value_counts().iloc[:10].index)
plt.title('Top 10 Nationalities')
plt.show()
```

```python
 #Bivariate Analysis
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='age', y='overall', hue='preferred_foot', alpha=0.6, data=df)
plt.title('Overall Rating vs Age by Footedness')
plt.show()
plt.figure(figsize=(10,8))
sns.heatmap(df[['age','overall','potential','value_eur','wage_eur']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
